```python
import requests
import re
import time
from bs4 import BeautifulSoup
```

```python
# Step 1: compile a list of links from the search url
head={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36'}
movie_links = []
for movie in film_list:
    clean_movie = re.sub(r"[^\w\d\s]","",movie)
    formatted_movie = clean_movie.replace(" ","%20")
    search_url = "https://www.imdb.com/search/title/?title=" + formatted_movie + "&title_type=feature"
    print(search_url)
    raw_html = requests.get(search_url, headers=head).content
    time.sleep(2)
    soup_doc = BeautifulSoup(raw_html, "html.parser")
    try:
        page_link_element = soup_doc.find(class_="sc-43986a27-0 gUQEVh").a
        page_link = page_link_element.get('href')
        movie_link_dic = {
            'movie': movie,
            'link': page_link
        }
        movie_links.append(movie_link_dic)
    except:
        print(movie + " did not work")
movie_links
#once you get the link, you want to put it in a list of dict: movie_name and link
```

```python
links_df = pd.DataFrame(movie_links)
links_df.to_csv("BBC_movie_links.csv")
```

```python
import json
movie_details = []
for movie in movie_links:
    link = movie['link']
    movie_name = movie['movie']
    full_link = "https://www.imdb.com/"+link
    raw_html = requests.get(full_link, headers=head).content
    # Process raw_html here as needed, you have the HTML content of each page
    time.sleep(2)
    soup_doc = BeautifulSoup(raw_html, "html.parser")
    # look for budget
    try:
        budget_element = soup_doc.find('li', attrs={'data-testid': 'title-boxoffice-budget'}).text
        regex_for_budget = r"(\d+,\d+,\d+[,\d+]?)"
        movie['budget'] = re.findall(regex_for_budget,budget_element)
        # look for gross_ww
        gross_ww_element = soup_doc.find('li', attrs={'data-testid': 'title-boxoffice-cumulativeworldwidegross'}).text
        regex_for_gross_ww = r"(\d+,\d+,\d+[,\d+]?)"
        movie['gross_ww'] = re.findall(regex_for_gross_ww,gross_ww_element)
        # look for runtime
        movie['runtime'] = soup_doc.find_all('li', class_='ipc-inline-list__item')[6].get_text(strip=True)
        # look for IMDb rating
        movie['rating'] = soup_doc.find_all('span', class_='sc-bde20123-1 cMEQkK') [0].get_text(strip=True)
        # find genre
        text_json = soup_doc.find_all('script')[2].text
        json_object = json.loads(text_json)
        movie['genre'] = json_object['genre']
        movie_details.append(movie)
    except:
        movie_details.append(movie)
```
